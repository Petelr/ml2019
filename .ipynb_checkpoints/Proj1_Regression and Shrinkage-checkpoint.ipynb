```python
w = np.array([-0.8, 2.1, 1.5])
b = 10
# loc=>mean scale=>standard diviation
eps = np.random.normal(loc=0, scale=np.sqrt(10) ,size=200)
# adding noise
y = w*np.transpose(x)+b
y[:,0] = y[:,0] + eps
y[:,1] = y[:,1] + eps
y[:,2] = y[:,2] + eps
# selecting first 100 sample as training data and Standardize them
train_x = x.T[0:100]
train_x = train_x.T
train_y = y[0:100]

test_x  = x.T[100:200]
text_x  = test_x.T
text_y  = y[100:200]
```

```python
from sklearn import preprocessing
# standardizing test and training data
train_x = preprocessing.scale(train_x)
train_y = preprocessing.scale(train_y)
```

```python
# setting up the lambda test array
lambda_array = np.arange(0,0.51,0.01)

def rigid_regression(x_train, y_train, _lambda):
    ones    = np.ones(len(x_train.T))
    x_train = np.column_stack((ones, x))
    
    lambda_identity = _lambda*np.identity(len(x_train.T))
    the_inverse     = np.linalg.inv(np.dot(x_train.T,x_train) + lambda_identity)
    w_hat           = np.dot(np.dot(the_inverse, x_train.T),y_train)
    
    return w_hat
```
